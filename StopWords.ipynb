```python
from goose3 import Goose
```

```python
texto = article.cleaned_text
print(texto)
```

```python
# Tokenização
from nltk.tokenize import word_tokenize

word_text = word_tokenize(texto,language = 'portuguese')
print(word_text)
```

```python
#Sentenças

from nltk.tokenize import sent_tokenize

sentencas_text = sent_tokenize(texto,language = 'portuguese')
print(sentencas_text)
```

```python
#Distribuição de frequência das palavras

from nltk.probability import FreqDist
fdisk_text = FreqDist(word_text)
fdisk_text
```

```python
fdisk_text.most_common(10)
```

```python
fdisk_text.plot(10)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwordsnltk = nltk.corpus.stopwords.words('portuguese')
print(stopwordsnltk)
```

```python
new_stopwords = ['As',',','...','.','?','!']
stopwordsnltk.extend(new_stopwords)
print(stopwordsnltk)
```

```python
new_words = []
for word in word_text:
    if word not in stopwordsnltk:
      new_words.append(word)
print(new_words)
```

```python
def plot_cloud(new_words):
      wordcloud = WordCloud(
      width=800,
      height=400,
      max_words=500,
      ).generate(' '.join(new_words))
      plt.figure(figsize=(10,10))
      plt.imshow(wordcloud,interpolation='bilinear')
      plt.axis('off')
      plt.show()


plot_cloud(new_words)
```
