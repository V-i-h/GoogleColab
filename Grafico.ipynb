```python
# prompt: pesquisar os top 10 paises com imigrantes do Canadá

top_10 = df.sort_values(by='Total', ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
ax = sns.barplot(data=top_10, y=top_10.index,x='Total',orient='h')
ax.set(
    title = 'Os dez paizes com mais imigrantes para o Canadá',
    xlabel = 'Total de imigrantes',
    ylabel = ''
)
plt.show()
```

```python
def gerar_grafico_paleta(pallete):
  fig,ax = plt.subplots(figsize = (8,4))
  ax = sns.barplot(data=top_10, y=top_10.index,x='Total',orient='h',palette=pallete)
  ax.set_title ('Os dez paises com mais imigrantes para o Canadá',loc='left',fontsize= 16)
  ax.set_xlabel ('Total de imigrantes',fontsize=14)
  ax.set_ylabel ('')
  plt.show()
```

```python
sns.set_theme(style="white")
gerar_grafico_paleta('tab10')
```

Gráficos Interativos com Plotly

```python
import plotly.express as px
```

```python
anos = list(map(str,range(1980,2014)))
```

```python
brasil = df.loc['Brasil',anos]
```

```python
brasil_dict = {'ano':brasil.index.tolist(),'imigrantes':brasil.values.tolist()}
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil.head()
```

```python
fig = px.line(dados_brasil,x='ano',y='imigrantes',title='Imigrantes do Brasil para o Canadá de 1980 a 2013')
fig.update_layout(width=1000,height=500,font_family='Arial',
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=22,xaxis_title='Ano',
                  yaxis_title='Número de Imigrantes')
fig.update_traces(line_color ='red',line_width=3)
fig.update_xaxes(tickangle=45)
fig.show()
```

```python
fig.write_html('imigrantes_brasil.html')
```
