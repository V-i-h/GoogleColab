# Parte 1 - Realize a tokenização de palavras e setenças


```python
texto = article.cleaned_text
```

```python
print(texto)
```

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
```

```python
word_texto = word_tokenize(texto, language ='portuguese')
print(word_texto)
```

```python
sentencas_texto = sent_tokenize(texto, language = 'portuguese')
print(sentencas_texto)
```

# Parte 3 - Crie uma função plot_cloud() para WordCloud





```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
def plot_cloud(word_texto):
  # Create a WordCloud object

  wordcloud = WordCloud(width=800, height=400, max_words=500).generate(" ".join(word_texto))
  # Generate the word cloud from the text data

  plt.figure(figsize=(8,10))
  plt.imshow(wordcloud, interpolation='bilinear') # Display the generated word cloud image
  plt.axis('off')
  plt.show()

plot_cloud(word_texto)
```
